```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img1 = cv2.imread('../DATA/Nadia_Murad.jpg',0)
img2 = cv2.imread('../DATA/Denis_Mukwege.jpg',0)
img3 = cv2.imread('../DATA/solvay_conference.jpg',0)
```

```python
#plt.imshow(img1)
#plt.imshow(img2)
plt.imshow(img3,cmap='gray')
```

```python
face_cascade=(cv2.CascadeClassifier('../DATA/haarcascades/haarcascade_frontalface_default.xml'))
```

```python
def detect_face(img):
    face_img =img.copy()
    
    face_rect =face_cascade.detectMultiScale(face_img,scaleFactor=1.2,minNeighbors=5)
    
    for (x,y,w,h) in face_rect:
        cv2.rectangle(face_img,(x,y),(x+w,y+h),(255,0,0),10)
        
    return face_img
```

```python
dec = detect_face(img3)
plt.imshow(dec,cmap='gray')
```

face detection from web cam

```python
cap=cv2.VideoCapture(0)

while True:
    
    ret,frame = cap.read(0)
    frame= detect_face(frame)
    cv2.imshow('video Face Detection',frame)
    
    c=cv2.waitKey(1)
    
    if c==27:
        break
cap.release()
cv2.destroyAllWindows()
```
