```python
import numpy as np
```

```python
from numpy import genfromtxt
data = genfromtxt('../DATA/bank_note_data.txt',delimiter=',')
```

```python
labels = data[:,4]
features = data[:,:4]
```

```python
X= features
y=labels
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=.33,random_state=42)
```

```python
from sklearn.preprocessing import MinMaxScaler
scalar_object = MinMaxScaler()
scalar_object.fit(X_train)

scalar_x = scalar_object.transform(X_train)
scalar_x1 = scalar_object.transform(X_test)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(4,input_dim=4,activation ='relu'))
model.add(Dense(8,activation ='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(scalar_x,y_train,epochs=50,verbose =2)
```

```python
model.predict_classes(scalar_x1)
```

```python
loss,accuracy = model.evaluate(x=scalar_x1,y=y_test)
print(loss)
print(accuracy)
```

```python
from sklearn.metrics import confusion_matrix,classification_report
predictions=model.predict_classes(scalar_x1)
confusion_matrix(y_test,predictions)
```

```python
print(classification_report(y_test,predictions))
```

```python
model.save('bank_note_solution.h5')
```
